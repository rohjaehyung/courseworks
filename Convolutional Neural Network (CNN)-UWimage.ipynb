```python
from PIL import Image
from pylab import *

im = Image.open('uw.png')
im_m = array(im)[50:299, :, :-1] #cropping whitespace
im_c = Image.fromarray(im_m)
imshow(im_c)
```

```python
from PIL import ImageFilter
import numpy as np
```

```python
blur = ImageFilter.Kernel((3,3), np.ones(9),9,0) #blurring the image
im_b = im_c.filter(blur)
imshow(im_b)
```

```python
laplace = ImageFilter.Kernel((3,3), (0,1,0, 1,-4,1, 0,1,0), 1, 0) # laplace predefined numbers
im_l = im_c.filter(laplace)
imshow(im_l)
```

```python
# Pooling
from skimage.measure import block_reduce
```

```python
im_pool_max = block_reduce(im_m, block_size = (6,6,1), func = np.max) #np.max makes Maximum pooling
imshow(im_pool_max)
```
